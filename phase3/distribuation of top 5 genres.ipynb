```python
import matplotlib.pyplot as plt 
import json 

with open('IMDB_crawled.json', 'r') as file:
    data = json.load(file)
top_n_genres = 5
number = {}
for movie in data :
    genres = movie['genres'] 
    for genre in genres : 
        if genre not in number.keys() : 
            number[genre] = 0 
        number[genre] += 1
arr = []
for genre in number.keys() :
    arr.append((genre,number[genre]))
arr = sorted(arr, key=lambda x:x[1], reverse=True)[0:top_n_genres] 
top_genres = [x[0] for x in arr]
indexing_top_genres = {}
for i, genre in enumerate(top_genres) : 
    indexing_top_genres[genre] = i
```

```python
dist = [number[genre] for genre in top_genres]
plt.bar(top_genres, dist)
```
